# 이미지 파일 열기

```python
import cv2

# 이미지 파일 열기
image = cv2.imread('./temp_image/siba.jpg')

# 1.이미지를 직접 윈도우에 표시
cv2.imshow('Image', image)

# 키 입력 대기
cv2.waitKey(0)

# 창 닫기
cv2.destroyAllWindows()
```

## ※추가: 이미지를 주피터 노트북의 출력셀에 표시하기

```python
import cv2
import matplotlib import pyplot as plt

# 이미지 파일 열기
image = cv2.imread('./temp_image/siba.jpg')

# OpenCV는 BGR로 이미지를 로드하지만, matplotlib는 RGB로 이미지를 표시합니다.
# 따라서 BGR에서 RGB로 이미지 색상 순서를 변경해야 합니다.
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 이미지 표시
plt.imshow(image_rgb)
plt.title('Image')
plt.show()
```
